```python
#Read and Display Image
```

```python
import cv2
from matplotlib import pyplot as plt

# Read image from file
img = cv2.imread(r"C:\Users\purva.nanivadekar\Downloads\TREE.jpg")  # Replace with your image filename

# Check if the image was loaded successfully
if img is not None:
    # Convert image from BGR (OpenCV default) to RGB for correct display in matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title('Loaded Image')
    plt.axis('off')
    plt.show()
else:
    print("Error: Image not found.")
```

```python
#Observe Image Properties
```

```python
# Print out image properties
print(f"Shape (Height, Width, Channels): {img.shape}")
print(f"Image Datatype: {img.dtype}")
```

```python
# Split the Layers (Channels)
```

```python
# Split the image into Blue, Green, and Red channels
b, g, r = cv2.split(img)

# Display each channel
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')
plt.show()
```

```python
#Convert to Grayscale
```

```python
# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
#Crop an Image
```

```python
# Define ROI coordinates (y1:y2, x1:x2)
cropped = img[50:200, 100:300]
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')
plt.show()
```

```python
#Arithmetic Operations (Image Blending)
```

```python
# Make sure two images are the same size
img2 = cv2.resize(img, (img.shape[1], img.shape[0]))  # For demonstration

# Weighted addition (blending)
blended = cv2.addWeighted(img, 0.2, img2, 0.3, 0)
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
plt.title('Blended Image')
plt.axis('off')
plt.show()
```

```python
#Logical Operations (on Binary Images)
```

```python
# Create binary images (using thresholding)
ret, binary1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
ret, binary2 = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY_INV)

# Logical AND
and_img = cv2.bitwise_and(binary1, binary2)
# Logical OR
or_img = cv2.bitwise_or(binary1, binary2)
# Logical XOR
xor_img = cv2.bitwise_xor(binary1, binary2)
# Logical NOT
not_img = cv2.bitwise_not(binary1)

# Display results
fig, axs = plt.subplots(1,4, figsize=(16,4))
titles = ['AND', 'OR', 'XOR', 'NOT']
images = [and_img, or_img, xor_img, not_img]
for i in range(4):
    axs[i].imshow(images[i], cmap='gray')
    axs[i].set_title(titles[i])
    axs[i].axis('off')
plt.show()
```
